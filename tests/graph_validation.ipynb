```python
import sys
sys.path.append('..')

import networkx as nx
import matplotlib.pyplot as plt
from models.networks.generators import get_graph

def get_graph_types():
    return [
        ('complete', {}),
        ('empty', {}),
        ('hypercube', {}),
        ('power', {'m': 2, 'p': 0.3}),
        ('random', {'p': 0.3}),
        ('ring_cliques', {}),
        ('small_world', {'k': 2, 'p': 0.3}),
        ('small_world', {'k': 2, 'p': 0.7}),
        ('small_world', {'k': 4, 'p': 0.3}),
        ('small_world', {'k': 4, 'p': 0.7}),
        ('star', {}),
        ('tree', {}),
        ('wheel', {}),
        ('windmill', {})
    ]

def plot_one_graph(graph_type, size, opts):
    try:
        # Generate graph using new interface
        G = get_graph(graph_type, size, **opts)
        
        # Create layout (with larger spacing for better visibility)
        layout = nx.kamada_kawai_layout(G)
        
        # Draw graph
        nx.draw(G, layout, 
                with_labels=True,
                node_color='lightblue',
                node_size=500,
                font_size=10,
                font_weight='bold')
        
        # Calculate and display metrics
        metrics = []
        if len(G) == size:
            metrics.append(f'Size: {len(G)}')
        else:
            metrics.append(f'WRONG SIZE: {len(G)}, not {size}')
        
        metrics.append(f'Edges: {len(G.edges)}')
        
        try:
            metrics.append(f'Density: {nx.density(G):.2f}')
        except:
            metrics.append('Density: N/A')
            
        try:
            metrics.append(f'Clustering: {nx.average_clustering(G):.2f}')
        except:
            metrics.append('Clustering: N/A')
            
        try:
            metrics.append(f'Connected: {nx.is_connected(G)}')
        except:
            metrics.append('Connected: N/A')
            
        # Add parameter info if any
        if opts:
            param_str = ', '.join(f'{k}={v}' for k, v in opts.items())
            metrics.append(f'Params: {param_str}')
        
        plt.title(f'{graph_type.capitalize()}\n' + '\n'.join(metrics))
        
    except Exception as e:
        plt.text(0.5, 0.5,
                 f'Error generating {graph_type}:\n{str(e)}',
                 ha='center',
                 va='center')
    
def plot_all_graphs(size):
    """Plot all graph types with given size."""
    graph_types = get_graph_types()
    
    # Set up subplot grid
    n_graphs = len(graph_types)
    n_cols = 3
    n_rows = (n_graphs + n_cols - 1) // n_cols
    
    fig = plt.figure(figsize=(15, 5*n_rows))
    
    for idx, (graph_type, opts) in enumerate(graph_types, 1):
        plt.subplot(n_rows, n_cols, idx)
        plot_one_graph(graph_type, size, opts)
    
    plt.suptitle(f'Graph Types with {size} Nodes', y=1.02, fontsize=16)
    plt.tight_layout()
    plt.show()

def plotter(size):
    """Run the plots with their message"""
    print(f"\nGenerating graphs with {size} nodes...")
    plot_all_graphs(size)
```

```python
# Test with all valid sizes (valid_sizes = [1, 2, 4, 8, 16, 32])
plotter(1)
```
